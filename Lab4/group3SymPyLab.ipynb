## SymPy’s Symbolic and Numercical Complex Evaluations

**My own example**

image.png

```
from sympy import symbols, I, N, pi, expand
x3, x4, y3, y4, z3, z4 = symbols("x3 x4 y3 y4 z3 z4", real=True)  
x = x3 + I*x4
y = y3 + I*y4
z = z3 + I*z4

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))

w=N(1/(pi + I), 20)
print('w=',w)
```
